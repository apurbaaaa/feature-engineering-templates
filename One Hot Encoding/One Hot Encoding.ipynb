### How do we prevent this from happening? ###
limit one hot encoding to the 10 most frequent labels of the variable (column). Meaning, only make one binary variable for each of the 10 most frequent labels only.

```python
top_10 = df.X2.value_counts().sort_values(ascending= False).head(10) #highest frequency values in descending order of column X2
```

```python
top_10 = [i for i in top_10.index]
```

```python
top_10
```

```python
for label in top_10:
    df[label] = np.where(df["X2"] == label , 1, 0) #wherever X2 matches with label, that is top 10 label, put 1 else 0
```

```python
df[["X2"]+top_10].head(20)
```

```python
def one_hot_top_X(df, variable, top_x_labels):
    for label in top_x_labels:
        df[variable+"_"+label] = np.where(df["X2"] == label , 1, 0)
```

```python
df = pd.read_csv("./train.csv", usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
```

```python
one_hot_top_X(df, 'X2', top_10)
```
